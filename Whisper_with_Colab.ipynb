```python
import re

with open("demo.txt") as f:
    text = f.read()

# simple sentence split
target_sentences = [s.strip() for s in re.split(r'[.!?]', text) if s.strip()]
print(f"Total sentences: {len(target_sentences)}")
```

```python
from sentence_transformers import SentenceTransformer, util

st_model = SentenceTransformer("all-MiniLM-L6-v2")

audio_sents = [seg["text"] for seg in result["segments"]]

emb_audio  = st_model.encode(audio_sents, convert_to_tensor=True)
emb_target = st_model.encode(target_sentences, convert_to_tensor=True)

matches = []
for i, a in enumerate(audio_sents):
    cos_sim = util.cos_sim(emb_audio[i], emb_target)[0]
    best_idx = int(cos_sim.argmax())
    matches.append({
        "start": result["segments"][i]["start"],
        "end": result["segments"][i]["end"],
        "audio_text": a,
        "matched_text": target_sentences[best_idx],
        "similarity": float(cos_sim[best_idx])
    })

print(f"Matched {len(matches)} segments.")
```

```python
import pandas as pd

df = pd.DataFrame(matches)
pd.set_option("display.max_colwidth", None)
df.head(10)
```
